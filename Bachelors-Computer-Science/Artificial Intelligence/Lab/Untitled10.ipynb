```python
import pandas as pd
import numpy as np
data=pd.DataFrame({'x1':np.random.normal(0,2,1000),'x2':np.random.normal(5,3,1000),'x3':np.random.normal(-5,5,1000)})
print(data.head())
data.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
stsc=StandardScaler()
stsc.fit(data)
data1=stsc.transform(data)
data1=pd.DataFrame(data1, columns=['x1','x2','x3'])
data1.plot.kde()
```

```python
data3=pd.DataFrame({'x1':np.random.chisquare(8,1000),'x2':np.random.beta(8,2,1000)*40,'x3':np.random.normal(-2,2,1000)})
data3.head()
data3.plot.kde()
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
data0=minmax.fit_transform(data3)
data01=pd.DataFrame(data0, columns=['x1','x2','x3'])
data01.plot.kde()
```
